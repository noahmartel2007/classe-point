## 1. Importation de la librairie
En Python, le calcul matriciel ne se fait pas nativement avec des listes (trop lent). On utilise **Numpy**, la référence pour le calcul scientifique.

Par convention, on l'importe sous l'alias `np`.

```python
import numpy as np

print("Numpy version :", np.__version__)
```

## 2. Création de matrices (Tableaux 2D)
Une matrice est un tableau de nombres organisé en lignes et en colonnes.
Pour créer une matrice $A = \begin{pmatrix} 1 & 2 & 3 \\ 4 & 5 & 6 \end{pmatrix}$, on utilise `np.array`.

```python
# Création d'une matrice A (2 lignes, 3 colonnes)
A = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

print("Matrice A :\n", A)

# Afficher les dimensions de la matrice (Lignes, Colonnes)
print("\nDimensions (Shape) de A :", A.shape)
```

### 2.1 Matrices particulières
En électronique ou en traitement d'image, on a souvent besoin de matrices vides ou remplies de 1.

```python
# Matrice de Zéros (ex: pour initialiser un buffer)
Z = np.zeros((3, 3))
print("Matrice de zéros :\n", Z)

# Matrice de Uns (ex: pour initialiser un calcul)
O = np.ones((3, 3))
print("Matrice de uns :\n", O)

# Matrice identité (diagonale à 1, le reste à 0)
# Très important pour l'inversion de matrice
I = np.eye(3)
print("\nMatrice Identité I3 :\n", I)

# Matrice de 5 (ex: pour initialiser un calcul)
Fives = np.ones((3, 3)) * 5
print("Matrice de 5 :\n", Fives)


```

## 3. Accéder aux éléments (Indexing & Slicing)
La syntaxe est : `Matrice[ligne, colonne]`. **Attention : les indices commencent à 0.**

```python
M = np.array([
    [10, 20, 30],
    [40, 50, 60],
    [70, 80, 90]
])

# 1. Accéder à la valeur 50 (ligne index 1, colonne index 1)
val = M[1, 1]
print("Élément central :", val)

# 2. Extraire la première ligne complète
ligne_1 = M[0, :]
print("Première ligne :", ligne_1)

# 3. Extraire la dernière colonne complète
col_fin = M[:, 2]
print("Dernière colonne :", col_fin)
```

## 4. Opérations Arithmétiques de base
Avec Numpy, les opérations `+`, `-`, `*`, `/` se font **élément par élément**.

Soient deux matrices :
$A = \begin{pmatrix} 1 & 2 \\ 3 & 4 \end{pmatrix}$ et $B = \begin{pmatrix} 10 & 10 \\ 10 & 10 \end{pmatrix}$

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[10, 10], [10, 10]])

# Addition
print("A + B =\n", A + B)

# Multiplication par un scalaire (ex: amplifier un signal par 2)
print("\n2 * A =\n", 2 * A)

# ATTENTION : Ceci est une multiplication élément par élément, PAS le produit matriciel
print("\nA * B (élément par élément) =\n", A * B)
```

Autre exemple avec une boucle

```python
dim = 5
tab = np.zeros((dim, dim))
for l in range(dim):
  for c in range(dim):
    tab[l, c] = (10 * l) + c

print(tab)
tab[3, :] += 100
print(tab)
```

## Synthèse de la première partie :
- Écrire une fonction stairs(k) qui crée un tableau carré de dimension 2k + 1
- Aux 4 coins du tableau, on trouve la valeur 0
- Au centre du tableau, on trouve la valeur 2k
- Quand on se déplace horizontalement ou verticalement d'une case vers le centre, la valeur augmente de 1

```python
print(stairs(3))
```

```python
import matplotlib.pyplot as plt

def stairs(k):
  dim = 2 * k + 1
  tab = np.zeros((dim,dim))
  for l in range(k+1):
    for c in range(k+1):
      tab[l, c]= (0 + l) + c
      tab[l, 2*k-c]= l + c
      tab[2*k-l, c] = l + c
      tab[2*k-l, 2*k-c] = l + c
  return tab
#print(stairs(4))
plt.imshow(stairs(100),cmap='gist_rainbow')
```

## 5. Le Produit Matriciel (Le vrai !)
En mathématiques, le produit de deux matrices (lignes $\times$ colonnes) ne se note pas `*` en Python, mais **`@`** (ou `np.dot`).

```python
# Produit matriciel A x B
# C11 = (1*10) + (2*10) = 30
# C12 = (1*10) + (2*10) = 30
# ...

C = A @ B
print("Produit Matriciel A @ B =\n", C)
```

## 6. Algèbre Linéaire : Transposée, Déterminant et Inverse
Ces outils sont fondamentaux pour résoudre des systèmes linéaires (Loi des mailles).

* **Transposée ($A^T$)** : On inverse lignes et colonnes.
* **Déterminant** : Si `det(A) = 0`, la matrice n'est pas inversible.
* **Inverse ($A^{-1}$)** : La matrice qui permet de "diviser" par A.

```python
A = np.array([[1, 2], [3, 4]])

# 1. Transposée
print("Transposée de A :\n", A.T)

# 2. Déterminant
det_A = np.linalg.det(A)
print(f"\nDéterminant de A : {det_A:.2f}")

# 3. Inverse
if det_A != 0:
    A_inv = np.linalg.inv(A)
    print("\nInverse de A :\n", A_inv)

    # Vérification : A @ A_inv doit donner la matrice Identité
    print("\nVérification (A @ A_inv) :\n", np.round(A @ A_inv))
else:
    print("Matrice non inversible")
```

## 7. Exercice d'application : Résolution d'un circuit électrique
On cherche à résoudre un système d'équations issu d'un circuit électrique (Lois de Kirchhoff).

Soit le système suivant :
$$
\begin{cases}
3x + 4y = 10 \\
2x - y = 3
\end{cases}
$$

On l'écrit sous la forme matricielle $A \cdot X = B$.

```python
# Définition des matrices
A_sys = np.array([[3, 4], [2, -1]])
B_sys = np.array([10, 3])

print("Matrice des coefficients A :\n", A_sys)
print("Matrice des résultats B :\n", B_sys)

# Résolution automatique
solution = np.linalg.solve(A_sys, B_sys)

print("\n--- RÉSULTAT ---")
print(f"x = {solution[0]}")
print(f"y = {solution[1]}")
```
