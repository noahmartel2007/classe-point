### 1 - Coder une classe **Point** contenant 2 coordonnées **x** et **y**, un **constructeur** qui initialise ces coordonnées à 0, une méthode **deplacer()** qui ajoute **deltaX** et **deltaY** respectivement à **x** et à **y**, et une méthode **afficher()** qui affiche les coordonnées x et y.

```python
class Point:
  def __init__(self):
    self.x = 0
    self.y = 0
  def deplacer(self, deltaX, deltaY):
    self.x += deltaX
    self.y = self.y + deltaY
  def afficher(self):
    print("x = ", self.x)
    print(f"y = {self.y}")
```

```python
A = Point()
A.deplacer(3, 2)
A.deplacer(3, 2)
A.afficher()
```

### 2 - Coder une classe **Segment** contenant 2 points **A** et **B**, un **constructeur** qui initialise ce segment avec 2 points de coordonnées (0, 0), une méthode **longueur()** qui renvoie la longueur du segment AB, et une méthode **milieu()** qui renvoie un Point avec les coordonnées du milieu du segment AB.

```python
import math

class Segment:
  def __init__(self, A = Point(), B = Point()):
    self.A = A
    self.B = B
  def longueur(self):
    res = (self.A.x - self.B.x)**2 + (self.A.y - self.B.y)**2
    return math.sqrt(res)
  def milieu(self):
    M = Point()
    M.x = (self.B.x + self.A.x) / 2
    M.y = (self.B.y + self.A.y) / 2
    return M
```

### 3 - Tester les classes **Point** et **Segment** en créant 2 points **A** et **B**, puis en déplaçant A vers **(3, 2)**, B vers **(7, 5)**, puis en affichant la longueur du segment **AB** et les coordonnées du point **C**, milieu de ce segment.

```python
A = Point()
B = Point()
A.deplacer(3, 2)
B.deplacer(7, 5)
#A.afficher()
#B.afficher()
AB = Segment(A, B)
print(f"Longueur de AB : {AB.longueur()}")
print(f"Milieu de AB : {AB.milieu().afficher()}")
```

### 5 - Coder une classe **Rectangle** contenant :
- 4 points **A**, **B**, **C** et **D**,
- un **constructeur** qui initialise ce rectangle avec 2 points diagonalement opposés **A** et **C**,
- une méthode **perimetre()** qui renvoie la somme des longueurs des segments AB, BC, CD et DA,
- une méthode **surface()** qui renvoie la superficie du rectangle ABCD.

Les segments AB et CD sont parallèles, ainsi que les segments BC et DA.

image.png

```python
class Rectangle:
  def __init__(self,  A = Point(), C = Point()):
    self.A = A
    self.C = C
    self.B = Point()
    self.B.deplacer(self.A.x, self.C.y)
    self.D = Point()
    self.D.deplacer(self.C.x, self.A.y)
    self.AB = Segment(self.A, self.B)
    self.BC = Segment(self.B, self.C)
  def perimetre(self):
    return 2 * (self.AB.longueur() + self.BC.longueur())
  def surface(self):
    return self.AB.longueur() * self.BC.longueur()
```

### 6 - Tester la classe **Rectangle**:

- en affichant la superficie de ce rectangle.

```python
A = Point()
C = Point()
A.deplacer(3, 2)
C.deplacer(7, 5)
ABCD = Rectangle(A, C)
#ABCD.D.afficher()
print(f"Périmètre de ABCD : {ABCD.perimetre()}")
print(f"Surface de ABCD : {ABCD.surface()}")
```
